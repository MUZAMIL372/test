```python
#matrix plot
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
var=np.linspace(1,10,40).reshape(4,10)
var
sns.heatmap(var)
```

```python
sns.heatmap(var, vmin=0, vmax=40, cmap='PuOr', annot=True)
plt.show()
```

```python
annot_kws = {"fontsize": 12, "color": "yellow"}

# Create a heatmap with corrected parameters
sns.heatmap(var, vmin=0, vmax=40, cmap='PuOr', annot=True, annot_kws=annot_kws, linewidths=12,cbar=False)
plt.show()
```
